```python
from chessboard import Chessboard
import copy

def move_start(move_counter):
    while True:
        chessboard.board_view()
        start = input("С какой позиции вы хотите сходить (например A1) или напишите exit если у вас не осталось ходов или вы сдаётесь: ")
        if start == 'exit':
            break
        if len(start) == 2 and start[0] in "ABCDEFGH" and start[1] in "12345678":
            if chessboard.get_piece_from_coordinates(*tuple(start)):
                if (move_counter % 2 != 0 and chessboard.get_piece_from_coordinates(*tuple(start)).color == 'white') or (move_counter % 2 == 0 and chessboard.get_piece_from_coordinates(*tuple(start)).color == 'black'):
                    break;
        print("Введите корректные данные ячейки поля на которой находится фигура!")
    return start
def move_end(start):
    while True:
        chessboard.board_view()
        print("Ваши варианты хода этой фигурой: ")
        print(chessboard.get_piece_from_coordinates(*tuple(start)).moves({'white': chessboard.coordinates_pieces('white'),
                                                                        'black': chessboard.coordinates_pieces('black')}))
        end = input("На какую позицию вы хотите сходить: ")
        if len(end) == 2 and end[0] in "ABCDEFGH" and end[1] in "12345678":
            if tuple(end) in chessboard.get_piece_from_coordinates(*tuple(start)).moves({'white': chessboard.coordinates_pieces('white'),
                                                                                         'black': chessboard.coordinates_pieces('black')}):
                break
        print("Введите корректные данные ячейки поля на которой находится фигура!")
    return end
def valid_move(start, end):
    white_pieces = copy.deepcopy(chessboard.white_pieces)
    black_pieces = copy.deepcopy(chessboard.black_pieces)

    color = chessboard.get_piece_from_coordinates(*tuple(start)).color
    chessboard.move(start, end)

    if not chessboard.check_of_check(color):
        return True
    print("Такой ход невозможен, потому что у вас шах")

    chessboard.white_pieces = copy.deepcopy(white_pieces)
    chessboard.black_pieces = copy.deepcopy(black_pieces)
    return False

chessboard = Chessboard()
move_counter = 1
GAME = True

while(True):
    print(f"Ход номер: {move_counter}")
    while True:
        start_position = move_start(move_counter)
        if start_position == 'exit':
            print(f'{"белые" if move_counter%2==0 else "черные"} выиграли!')
            exit()
        end_position = move_end(start_position)
        if valid_move(start_position, end_position):
            break
    move_counter += 1




```
