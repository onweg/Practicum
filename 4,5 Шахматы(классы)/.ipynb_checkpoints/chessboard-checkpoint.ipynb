```python
from piece import Pawn, Horse, Elephant, Queen, King, Rook
import copy
class Chessboard:
    def __init__(self):
        self.black_pieces = [
            King('black', 'E', '8'),
            Queen('black', 'D', '8'),
            Elephant('black', 'F', '8'),
            Elephant('black', 'C', '8'),
            Horse('black', 'G', '8'),
            Horse('black', 'B', '8'),
            Rook('black', 'H', '8'),
            Rook('black', 'A', '8'),
            Pawn('black', 'A', '7'),
            Pawn('black', 'B', '7'),
            Pawn('black', 'C', '7'),
            Pawn('black', 'D', '7'),
            Pawn('black', 'E', '7'),
            Pawn('black', 'F', '7'),
            Pawn('black', 'G', '7'),
            Pawn('black', 'H', '7'),
        ]
        self.white_pieces = [
            King('white', 'E', '1'),
            Queen('white', 'D', '1'),
            Elephant('white', 'F', '1'),
            Elephant('white', 'C', '1'),
            Horse('white', 'G', '1'),
            Horse('white', 'B', '1'),
            Rook('white', 'H', '1'),
            Rook('white', 'A', '1'),
            Pawn('white', 'A', '2'),
            Pawn('white', 'B', '2'),
            Pawn('white', 'C', '2'),
            Pawn('white', 'D', '2'),
            Pawn('white', 'E', '2'),
            Pawn('white', 'F', '2'),
            Pawn('white', 'G', '2'),
            Pawn('white', 'H', '2'),
        ]

    def board_view(self):
        print('    A B C D E F G H')
        print()
        for j in "87654321":
            print(j, end='   ')
            for i in "ABCDEFGH":
                if self.get_piece_from_coordinates(i, j):
                    print(self.get_piece_from_coordinates(i, j).symbol, end=' ')
                else:
                    print('.', end=' ')


            print('  ' + j)
        print('\n    A B C D E F G H')

    def get_piece_from_coordinates(self, let, num):
        for piece in self.black_pieces:
            if (let, num) == (piece.let, piece.num):
                return piece
        for piece in self.white_pieces:
            if (let, num) == (piece.let, piece.num):
                return piece
        return None

    def coordinates_pieces(self, color = ""):
        if color == "white":
            return [(piece.let, piece.num) for piece in self.white_pieces]
        if color == "black":
            return [(piece.let, piece.num) for piece in self.black_pieces]
        return [(piece.let, piece.num) for piece in self.white_pieces] + [(piece.let, piece.num) for piece in self.black_pieces]

    def move(self, start, end):
        if self.get_piece_from_coordinates(*tuple(end)):
            for piece in self.white_pieces:
                if tuple(end) == (piece.let, piece.num):
                    self.white_pieces.remove(piece)
                    break
            for piece in self.black_pieces:
                if tuple(end) == (piece.let, piece.num):
                    self.black_pieces.remove(piece)
                    break
        for index in range(len(self.white_pieces)):
            if start[0] == self.white_pieces[index].let and start[1] == self.white_pieces[index].num:
                self.white_pieces[index].let = end[0]
                self.white_pieces[index].num = end[1]
                break
        for index in range(len(self.black_pieces)):
            if start[0] == self.black_pieces[index].let and start[1] == self.black_pieces[index].num:
                self.black_pieces[index].let = end[0]
                self.black_pieces[index].num = end[1]
                break

    def figures_who_are_in_danger(self, color):
        dangerous_cells_list = []
        all_coordinates = self.coordinates_pieces()

        if color == 'black':
            for piece in self.white_pieces:
                for let, num in piece.moves({'white': self.coordinates_pieces('white'),
                                             'black': self.coordinates_pieces('black')}):
                    if self.get_piece_from_coordinates(let, num):
                        dangerous_cells_list.append((let, num))
        else:
            for piece in self.black_pieces:
                for let, num in piece.moves({'white': self.coordinates_pieces('white'),
                                             'black': self.coordinates_pieces('black')}):
                    if self.get_piece_from_coordinates(let, num):
                        dangerous_cells_list.append((let, num))

        return dangerous_cells_list

    def check_of_check(self, color):
        for piece in self.white_pieces+self.black_pieces:
            if piece.symbol in 'Kk' and piece.color == color:
                if (piece.let, piece.num) in self.figures_who_are_in_danger(color):
                    return 'have check'
                break
        return ""

    # def checkmate_check(self):
    #     move_list = []
    #
    #     if self.check_of_check() == "white have check":
    #         for piece in self.white_pieces:
    #             for let, num in piece.moves(self.get_piece_from_coordinates(piece.let, piece.num).moves({'white':self.coordinates_pieces('white'),
    #                                                                                                      'black':self.coordinates_pieces('black')})):
    #                 white_piece_copy = copy.deepcopy(self.white_pieces)
    #                 black_piece_copy = copy.deepcopy(self.black_pieces)
    #                 #поменяли фигуру
    #                 #проверили на шах
    #                     #проверить состоятл ли координаты короля в красных клетках
    #                         #перебрать все черные и проверить на in
    #                 for index_copy in range(len(white_piece_copy)):
    #                     if (white_piece_copy[index_copy].let, white_piece_copy[index_copy].num) == (piece.let, piece.num):
    #                         (white_piece_copy[index_copy].let, white_piece_copy[index_copy].num) = (let, num)
    #                         break
    #
    #                 for black_piece in black_piece_copy:
    #                     if
    #
    #
    #                 for piece_copy in white_piece_copy:
    #                     if piece_copy.symbol == 'K':
    #                         break
    #
    #                 for piece in self.white_pieces:
    #                     if piece.symbol == 'K':
    #                         if (piece.let, piece.num) in self.figures_who_are_in_danger('white'):
    #                             return 'white have check'
    #                         break
    #
    #                 #проверить эту растановку на шах



```
